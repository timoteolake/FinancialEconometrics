```python
import numpy as np
import pandas as pd 

import pandas_datareader.data as web

# настройки визуализации
import matplotlib.pyplot as plt

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
rates = web.DataReader(name='FEDFUNDS', data_source='fred', start='1990-1')
rates
```

```python
rate1 = web.DataReader(name='FEDFUNDS', data_source='fred', start='1990-1')
rate1.index = rate1.index.to_period(freq='M')
rate1
```

```python
rate2 = web.DataReader(name='FEDFUNDS', data_source='fred', start='1990-1')
rate2.index = rate2.index.to_period(freq='10Y')
rate2
```

```python
rate1.plot()
```

```python
rate2.plot()
```

```python
rate1.plot(label='Quarter', color='b')
rate2.plot(label='Decade', color='r')
plt.legend()
plt.title("Quarter and Decade Rates")
plt.show()
```

```python
rate1.diff()
```

```python
rate2.diff()
```

```python
ax = rate1.diff().plot(title='diff(log(M2)')
# надпись по ос oX
ax.set_xlabel('Date')
# надпись по ос oY
ax.set_ylabel('US M2')
# отобразить сетку
ax.grid()
# удалим легенду
ax.legend().remove()
bx = rate2.diff().plot(title='diff(log(M2)')
# надпись по ос oX
bx.set_xlabel('Date')
# надпись по ос oY
bx.set_ylabel('US M2')
# отобразить сетку
bx.grid()
# удалим легенду
bx.legend().remove()

plt.show()
```

```python
ax = rate1.diff().diff().plot(title='diff(diff(log(M2))')
# надпись по ос oX
ax.set_xlabel('Date')
# надпись по ос oY
ax.set_ylabel('US M2')
# отобразить сетку
ax.grid()
# удалим легенду
ax.legend().remove()
bx = rate2.diff().diff().plot(title='diff(diff(log(M2))')
# надпись по ос oX
bx.set_xlabel('Date')
# надпись по ос oY
bx.set_ylabel('US M2')
# отобразить сетку
bx.grid()
# удалим легенду
bx.legend().remove()

plt.show()
```
